```python
%matplotlib inline
import matplotlib.pyplot as pyplot
import numpy
```

```python
import time
class wallclock:
    def __init__(self):
        self.begin = time.time()
    def elapsed(self):
        return time.time() - self.begin
```

# Bottom-Up Segmentation

```python
import slycat.timeseries.segmentation
reload(slycat.timeseries.segmentation)

for times in [numpy.linspace(0, numpy.pi * 2, count) for count in [10, 100, 1000, 10000, 100000]]:
    numpy.random.seed(1234)
    coefficients = numpy.random.random(8) * 16
    values = numpy.sum(numpy.column_stack([numpy.sin(k * times) / k for k in coefficients]), axis=1)
    
    timer = wallclock()
    e_times, e_values = slycat.timeseries.segmentation.bottom_up(times, values, max_error=0.1)
    elapsed = timer.elapsed()
    
    pyplot.figure()
    pyplot.plot(times, values, color="black", alpha=0.2, linewidth=4.0)
    pyplot.plot(e_times, e_values, color="black", linewidth=2.0)
    pyplot.title("Reduced %s samples to %s in %.3fs" % (len(times), len(e_times), elapsed))
    pyplot.gcf().set_size_inches(16, 6)
```
