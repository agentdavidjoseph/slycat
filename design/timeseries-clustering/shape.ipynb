```python
import numpy
import toyplot
```

```python
numpy.random.seed(1234)
t = numpy.linspace(0, 1, 1000)
waves = [numpy.sum([numpy.sin(k * t) for k in (50 * numpy.random.random(4))], axis=0) for i in range(20)]
```

```python
canvas = toyplot.Canvas()
axes = canvas.axes()
for wave in waves:
    axes.plot(t, wave)
```
