```python
import numpy
import toyplot
```

```python
windows = [
    ("Bartlett", numpy.bartlett(1000)),
    ("Blackman", numpy.blackman(1000)),
    ("Hamming", numpy.hamming(1000)),
    ("Hanning", numpy.hanning(1000)),
    ("Kaiser 0", numpy.kaiser(1000, beta=0)),
    ("Kaiser 1", numpy.kaiser(1000, beta=1)),
    ("Kaiser 5", numpy.kaiser(1000, beta=5)),
    ("Kaiser 10", numpy.kaiser(1000, beta=10)),
    ("Kaiser 15", numpy.kaiser(1000, beta=15)),
    ]
```

```python
canvas = toyplot.Canvas()
for index, (label, window) in enumerate(windows):
    axes = canvas.axes(grid=(3, 3, index), gutter=40, ylabel=label)
    axes.plot(window)
```
