```python
import numpy
import scipy.signal
import scipy.spatial.distance
import toyplot
```

```python
t = numpy.linspace(0, 1, 1000)
numpy.random.seed(12)
timeseries = []
for i in range(20):
    frequency = numpy.random.uniform(1, 100)
    pulse_times = numpy.arange(0, 1, 1.0 / frequency)[1:]
    signal = numpy.sum([scipy.signal.gausspulse(t - pulse_time, 500, retenv=True)[1] for pulse_time in pulse_times], axis=0)
    timeseries.append(signal)
timeseries = numpy.array(timeseries)
```

```python
canvas = toyplot.Canvas(1000, 1000)
axes = canvas.axes(xlabel="Time")
axes.y(show=False)
for index, y in enumerate(timeseries):
    axes.plot(t, (index*2) + y, style={"stroke":"#343434"})
```

```python
canvas = toyplot.Canvas(300, 300)
axes = canvas.axes(xlabel="Distance", ylabel="Count")
axes.barplot(numpy.histogram(scipy.spatial.distance.pdist(timeseries, metric="euclidean")))
```

```python
import sklearn.cluster
kmeans = sklearn.cluster.KMeans(8)
kmeans.fit(timeseries)
clusters = kmeans.predict(timeseries)
cluster_order = numpy.argsort(clusters)
colormap = toyplot.color.CategoricalMap(toyplot.color.Palette())
```

```python
canvas = toyplot.Canvas(1000, 1000)
axes = canvas.axes(xlabel="Time")
axes.y(show=False)
for index, (cluster, y) in enumerate(zip(clusters[cluster_order], timeseries[cluster_order])):
    axes.plot(t, (2 * index) + y, style={"stroke":colormap.css(cluster)})
```
