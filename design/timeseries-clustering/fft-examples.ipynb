```python
from __future__ import division
```

```python
import numpy.fft
import toyplot
```

```python
t = numpy.linspace(0, 1, 200)
y = numpy.sum([amplitude * numpy.sin(t * 2 * numpy.pi * frequency) for frequency, amplitude in [(1.6, 1)]], axis=0)
n = len(t)
```

```python
canvas = toyplot.Canvas(500, 300)
axes = canvas.axes()
axes.plot(t, y)
```

```python
amplitudes = numpy.fft.fftshift(numpy.fft.fft(y) / n)
frequencies = numpy.fft.fftshift(numpy.fft.fftfreq(n, 1 / n))
```

```python
canvas = toyplot.Canvas(800)
axes = canvas.axes(xmin=-10, xmax=10)
axes.plot(frequencies, numpy.real(amplitudes))
axes.scatterplot(frequencies, numpy.real(amplitudes))
axes.plot(frequencies, -numpy.imag(amplitudes))
axes.scatterplot(frequencies, -numpy.imag(amplitudes))
```

```python
y2 = numpy.zeros(len(t))
for amplitude, frequency in zip(numpy.real(amplitudes), frequencies):
    y2 += amplitude * numpy.cos(t * 2 * numpy.pi * frequency)
for amplitude, frequency in zip(numpy.imag(amplitudes), frequencies):
    y2 += -amplitude * numpy.sin(t * 2 * numpy.pi * frequency)

canvas = toyplot.Canvas(500, 300)
axes = canvas.axes()
axes.plot(t, y2)
```
