```python
import numpy
import scipy.signal
import scipy.spatial.distance
import toyplot
```

```python
t = numpy.linspace(0, 1, 1000)
numpy.random.seed(12)
timeseries = numpy.array([scipy.signal.gausspulse(t - numpy.random.random(), 500, retenv=True)[1] for i in range(20)])
```

```python
canvas = toyplot.Canvas(1000, 500)
axes = canvas.axes()
for y in timeseries:
    axes.plot(t, y)
```

```python
D = scipy.spatial.distance.pdist(timeseries, metric="euclidean")
```

```python
canvas = toyplot.Canvas()
axes = canvas.axes()
axes.barplot(numpy.histogram(D))
```
