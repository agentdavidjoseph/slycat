```python
import numpy
import scipy.signal
import scipy.spatial.distance
import toyplot
```

```python
t = numpy.linspace(-1, 1, 2000)
numpy.random.seed(12)
timeseries = numpy.array([scipy.signal.gausspulse(t, numpy.random.uniform(5, 100), retenv=True)[1] for i in range(20)])
```

```python
canvas = toyplot.Canvas(500, 1000)
axes = canvas.axes(xlabel="Time")
axes.y(show=False)
for index, y in enumerate(timeseries):
    axes.plot(t, (index*2) + y, style={"stroke":"#343434"})
```

```python
canvas = toyplot.Canvas(800, 300)
axes = canvas.axes(xlabel="Time")
for index, y in enumerate(timeseries):
    axes.plot(t, y, style={"stroke":"#343434"})
```

```python
canvas = toyplot.Canvas(300, 300)
axes = canvas.axes(xlabel="Distance", ylabel="Count")
axes.barplot(numpy.histogram(scipy.spatial.distance.pdist(timeseries, metric="euclidean")))
```

```python
import sklearn.cluster
kmeans = sklearn.cluster.KMeans(5)
kmeans.fit(timeseries)
clusters = kmeans.predict(timeseries)
colormap = toyplot.color.CategoricalMap(toyplot.color.Palette())
```

```python
canvas = toyplot.Canvas(800, 300)
axes = canvas.axes(xlabel="Time")
for cluster, y in zip(clusters, timeseries):
    axes.plot(t, y, style={"stroke":colormap.css(cluster)})
```
