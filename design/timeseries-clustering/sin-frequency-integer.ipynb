```python
from __future__ import division
```

```python
import numpy.fft
import scipy.signal
import scipy.spatial.distance
import toyplot
```

# Data Synthesis

We begin by synthesizing our test data - in this case, a collection of sin waves, each with an integer number of waves:

```python
t = numpy.linspace(0, 1, 1000)
n = len(t)
numpy.random.seed(12)
timeseries = numpy.row_stack([numpy.sin(t * 2 * numpy.pi * (i+1)) for i in range(20)])
```

```python
canvas = toyplot.Canvas(1000, 1000)
axes = canvas.axes(xlabel="Time")
axes.y(show=False)
for index, y in enumerate(timeseries):
    axes.plot(t, (index*3) + y, style={"stroke":"#343434"})
```

# K-Means (Euclidean) Clustering on the Raw Timeseries

First, we treat the timeseries values as features, and perform K-Means clustering.

```python
features = timeseries
```

```python
canvas = toyplot.Canvas(300, 300)
axes = canvas.axes(xlabel="Distance", ylabel="Count")
axes.barplot(numpy.histogram(scipy.spatial.distance.pdist(features, metric="euclidean")))
```

```python
import sklearn.cluster
kmeans = sklearn.cluster.KMeans(8)
kmeans.fit(features)
clusters = kmeans.predict(features)
cluster_order = numpy.argsort(clusters)
colormap = toyplot.color.CategoricalMap(toyplot.color.Palette())
```

```python
canvas = toyplot.Canvas(1000, 1000)
axes = canvas.axes(xlabel="Time")
axes.y(show=False)
for index, (cluster, y) in enumerate(zip(clusters[cluster_order], timeseries[cluster_order])):
    axes.plot(t, (3 * index) + y, style={"stroke":colormap.css(cluster)})
```

# K-Means (Euclidean) Clustering on Time Domain Features

```python
amplitudes = numpy.fft.fftshift(numpy.fft.fft(timeseries, axis=1)/n, axes=1)
frequencies = numpy.fft.fftshift(numpy.fft.fftfreq(n, 1/n))

sin_features = numpy.abs(frequencies[numpy.argsort(numpy.abs(numpy.imag(amplitudes)), axis=1)[:,-1:]])
cos_features = numpy.abs(frequencies[numpy.argsort(numpy.abs(numpy.real(amplitudes)), axis=1)[:,-1:]])

features = numpy.column_stack((sin_features, cos_features))
```

```python
canvas = toyplot.Canvas(300, 300)
axes = canvas.axes(xlabel="Distance", ylabel="Count")
axes.barplot(numpy.histogram(scipy.spatial.distance.pdist(features, metric="euclidean")))
```

```python
import sklearn.cluster
kmeans = sklearn.cluster.KMeans(4)
kmeans.fit(features)
clusters = kmeans.predict(features)
cluster_order = numpy.argsort(clusters)
colormap = toyplot.color.CategoricalMap(toyplot.color.Palette())
```

```python
canvas = toyplot.Canvas(1000, 1000)
axes = canvas.axes(xlabel="Time")
axes.y(show=False)
for index, (cluster, y) in enumerate(zip(clusters[cluster_order], timeseries[cluster_order])):
    axes.plot(t, (3 * index) + y, style={"stroke":colormap.css(cluster)})
```
