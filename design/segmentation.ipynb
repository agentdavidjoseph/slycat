```python
%matplotlib inline
import scipy.signal
import matplotlib
import matplotlib.pyplot as pyplot

from slycat.timeseries.segmentation import sum_of_squares, bottom_up

matplotlib.rc("lines", markersize=5.0)
dark_a = "blue"
dark_b = "green"
dark_c = "red"
dark_d = "orange"
light_a = (0.7, 0.7, 1.0)
light_b = (0.7, 1.0, 0.7)
light_c = (1.0, 0.7, 0.7)
light_d = (1.0, 0.9, 0.7)
light_bin = (0.8, 0.8, 0.8)
```

Create timeseries $A$, $B$, $C$, and $D$ which have different start and end times and nonuniform sampling:

```python
import numpy
a_times = numpy.linspace(0, numpy.pi * 2, 100) ** 3
a_values = numpy.sin(a_times * 0.05)

b_times = numpy.linspace(0, numpy.pi * 1.2, 37) ** 4 + 20
b_values = numpy.sin(b_times * 0.017)

c_times = numpy.linspace(-1, 1, 23)
c_values = scipy.signal.gausspulse(c_times)
c_times = c_times * 100.0 + 121.5

d_times = numpy.linspace(50, 200, 531)
d_values = numpy.sin(d_times * 0.62) * 0.4 + 0.5
```

Plot the timeseries for reference:

```python
pyplot.plot(a_times, a_values, "-o", color=dark_a, label="$A$")
pyplot.plot(b_times, b_values, "-o", color=dark_b, label="$B$")
pyplot.plot(c_times, c_values, "-o", color=dark_c, label="$C$")
pyplot.plot(d_times, d_values, "-o", color=dark_d, label="$D$")
pyplot.legend(loc="best")
pyplot.gcf().set_size_inches(10, 8)
```

## Sliding Window Segmentation

Starting with the first point in each timeseries, a segment is grown until it exceeds some error bound.  The process repeats with the next data point not included in the newly approximated segment.

With this approach, the caller can specify the maximum error of any segment, but not the total error for the entire timeseries, nor the total number of segments.

```python
def sliding_window(times, values, maximum_error):
    anchors = numpy.zeros(len(times), dtype="bool")
    anchors[0] = True
    anchor = 0
    i = 2
    while i < len(times):
        if sum_of_squares(times[anchor:i+1], values[anchor:i+1]) > maximum_error:
            anchor = i - 1
            anchors[anchor] = True
        i += 1
    anchors[-1] = True
    return times[anchors], values[anchors]

error = 0.01
as_times, as_values = sliding_window(a_times, a_values, error)
bs_times, bs_values = sliding_window(b_times, b_values, error)
cs_times, cs_values = sliding_window(c_times, c_values, error)
ds_times, ds_values = sliding_window(d_times, d_values, error)

print "A Compression:", len(a_times), "=>", len(as_times)
print "B Compression:", len(b_times), "=>", len(bs_times)
print "C Compression:", len(c_times), "=>", len(cs_times)
print "D Compression:", len(d_times), "=>", len(ds_times)
```

Note that the sliding window segmentation does a poor job of capturing the peaks and valleys for time series $D'$, and a reasonable job on the other time series:

```python
pyplot.plot(a_times, a_values, "-o", color=light_a, markeredgecolor=light_a)
pyplot.plot(b_times, b_values, "-o", color=light_b, markeredgecolor=light_b)
pyplot.plot(c_times, c_values, "-o", color=light_c, markeredgecolor=light_c)
pyplot.plot(d_times, d_values, "-o", color=light_d, markeredgecolor=light_d)

pyplot.plot(as_times, as_values, "-o", color=dark_a, label="$A'$")
pyplot.plot(bs_times, bs_values, "-o", color=dark_b, label="$B'$")
pyplot.plot(cs_times, cs_values, "-o", color=dark_c, label="$C'$")
pyplot.plot(ds_times, ds_values, "-o", color=dark_d, label="$D'$")

pyplot.legend(loc="best")
pyplot.gcf().set_size_inches(10, 8)
```

## Top-Down Segmentation

A timeseries is recursively partitioned until some stopping criteria is met.

With this approach, the caller can specify the maximum error of any segment, the maximum error of the timeseries as a whole, or the desired number of segments.

```python
def top_down(times, values, maximum_error):
    def split(times, values, maximum_error, left, right, anchors):
        if sum_of_squares(times[left:right], values[left:right]) < maximum_error:
            return
        errors = numpy.array([sum_of_squares(times[left:i+1], values[left:i+1]) + sum_of_squares(times[i:right], values[i:right]) for i in range(left + 2, right - 2)])
        if not len(errors):
            return
        split_point = left + 2 + numpy.argmin(errors)
        anchors[split_point] = True
        split(times, values, maximum_error, left, split_point + 1, anchors)
        split(times, values, maximum_error, split_point, right, anchors)
    
    anchors = numpy.zeros(len(times), dtype="bool")
    anchors[0] = True
    anchors[-1] = True
    split(times, values, maximum_error, 0, len(times), anchors)

    return times[anchors], values[anchors]

error = 0.01
at_times, at_values = top_down(a_times, a_values, error)
bt_times, bt_values = top_down(b_times, b_values, error)
ct_times, ct_values = top_down(c_times, c_values, error)
dt_times, dt_values = top_down(d_times, d_values, error)

print "A Compression:", len(a_times), "=>", len(at_times)
print "B Compression:", len(b_times), "=>", len(bt_times)
print "C Compression:", len(c_times), "=>", len(ct_times)
print "D Compression:", len(d_times), "=>", len(dt_times)
```

However, note for timeseries $C'$ that this approach has completely obliterated the spike:

```python
pyplot.plot(a_times, a_values, "-o", color=light_a, markeredgecolor=light_a)
pyplot.plot(b_times, b_values, "-o", color=light_b, markeredgecolor=light_b)
pyplot.plot(c_times, c_values, "-o", color=light_c, markeredgecolor=light_c)
pyplot.plot(d_times, d_values, "-o", color=light_d, markeredgecolor=light_d)

pyplot.plot(at_times, at_values, "-o", color=dark_a, label="$A'$")
pyplot.plot(bt_times, bt_values, "-o", color=dark_b, label="$B'$")
pyplot.plot(ct_times, ct_values, "-o", color=dark_c, label="$C'$")
pyplot.plot(dt_times, dt_values, "-o", color=dark_d, label="$D'$")

pyplot.legend(loc="best")
pyplot.gcf().set_size_inches(10, 8)
```

## Bottom-Up Segmentation

Starting from the finest possible approximation, segments are merged until some stopping criteria is met.


```python
error = 0.01
ab_times, ab_values = bottom_up(a_times, a_values, error)
bb_times, bb_values = bottom_up(b_times, b_values, error)
cb_times, cb_values = bottom_up(c_times, c_values, error)
db_times, db_values = bottom_up(d_times, d_values, error)

print "A Compression:", len(a_times), "=>", len(ab_times)
print "B Compression:", len(b_times), "=>", len(bb_times)
print "C Compression:", len(c_times), "=>", len(cb_times)
print "D Compression:", len(d_times), "=>", len(db_times)
```

Bottom up segmentation achieves good results for every time series, with none of the problems identified for the other algorithms:

```python
pyplot.plot(a_times, a_values, "-o", color=light_a, markeredgecolor=light_a)
pyplot.plot(b_times, b_values, "-o", color=light_b, markeredgecolor=light_b)
pyplot.plot(c_times, c_values, "-o", color=light_c, markeredgecolor=light_c)
pyplot.plot(d_times, d_values, "-o", color=light_d, markeredgecolor=light_d)

pyplot.plot(ab_times, ab_values, "-o", color=dark_a, label="$A'$")
pyplot.plot(bb_times, bb_values, "-o", color=dark_b, label="$B'$")
pyplot.plot(cb_times, cb_values, "-o", color=dark_c, label="$C'$")
pyplot.plot(db_times, db_values, "-o", color=dark_d, label="$D'$")

pyplot.legend(loc="best")
pyplot.gcf().set_size_inches(10, 8)
```
