```python
%matplotlib inline
import scipy.signal
import matplotlib
import matplotlib.pyplot as pyplot

matplotlib.rc("lines", markersize=5.0)
dark_a = "blue"
dark_b = "green"
dark_c = "red"
dark_d = "orange"
light_a = (0.7, 0.7, 1.0)
light_b = (0.7, 1.0, 0.7)
light_c = (1.0, 0.7, 0.7)
light_d = (1.0, 0.9, 0.7)
light_bin = (0.8, 0.8, 0.8)
```

Create timeseries $A$, $B$, $C$, and $D$ which have different start and end times and nonuniform sampling:

```python
import numpy
a_times = numpy.linspace(0, numpy.pi * 2, 100) ** 3
a_values = numpy.sin(a_times * 0.05)

b_times = numpy.linspace(0, numpy.pi * 1.2, 37) ** 4 + 20
b_values = numpy.sin(b_times * 0.017)

c_times = numpy.linspace(-1, 1, 23)
c_values = scipy.signal.gausspulse(c_times)
c_times = c_times * 100.0 + 121.5

d_times = numpy.linspace(50, 200, 531)
d_values = numpy.sin(d_times * 0.62) * 0.4 + 0.5
```

Define our resampling bins, using the extents of timeseries _a_ because it has the earliest start time and latest end time:

```python
bin_times, bin_width = numpy.linspace(a_times.min(), a_times.max(), 20, retstep=True)
bin_edges = numpy.append(bin_times - (bin_width / 2), [bin_times[-1] + (bin_width / 2)])

pyplot.vlines(bin_edges, -1, 1, color=light_bin)
pyplot.plot(a_times, a_values, "-o", color=dark_a, label="$A$")
pyplot.plot(b_times, b_values, "-o", color=dark_b, label="$B$")
pyplot.plot(c_times, c_values, "-o", color=dark_c, label="$C$")
pyplot.plot(d_times, d_values, "-o", color=dark_d, label="$D$")
pyplot.legend(loc="best")
pyplot.gcf().set_size_inches(10, 8)
```

A piecewise linear approximation (PLA) uses linear interpolation at each sample point and yields good results, so long as the bin width is relatively small compared to features in the data - but note the change in spike amplitude in $C'$ and aliasing problems with $D'$.  The scipy code we use here returns NaN for times outside the original timeseries domain:

```python
import scipy.interpolate
pyplot.vlines(bin_edges, -1, 1, color=light_bin)
pyplot.plot(a_times, a_values, "-o", color=light_a, markeredgecolor=light_a)
pyplot.plot(b_times, b_values, "-o", color=light_b, markeredgecolor=light_b)
pyplot.plot(c_times, c_values, "-o", color=light_c, markeredgecolor=light_c)
pyplot.plot(d_times, d_values, "-o", color=light_d, markeredgecolor=light_d)

pyplot.plot(bin_times, scipy.interpolate.griddata(a_times, a_values, bin_times), "-o", color=dark_a, label="$A'$")
pyplot.plot(bin_times, scipy.interpolate.griddata(b_times, b_values, bin_times), "-o", color=dark_b, label="$B'$")
pyplot.plot(bin_times, scipy.interpolate.griddata(c_times, c_values, bin_times), "-o", color=dark_c, label="$C'$")
pyplot.plot(bin_times, scipy.interpolate.griddata(d_times, d_values, bin_times), "-o", color=dark_d, label="$D'$")
pyplot.legend(loc="best")
pyplot.gcf().set_size_inches(10, 8)
```

An alternate PLA implementation produces similar results, but duplicates the first or last value of a series outside its domain:

```python
pyplot.vlines(bin_edges, -1, 1, color=light_bin)
pyplot.plot(a_times, a_values, "-o", color=light_a, markeredgecolor=light_a)
pyplot.plot(b_times, b_values, "-o", color=light_b, markeredgecolor=light_b)
pyplot.plot(c_times, c_values, "-o", color=light_c, markeredgecolor=light_c)
pyplot.plot(d_times, d_values, "-o", color=light_d, markeredgecolor=light_d)

pyplot.plot(bin_times, numpy.interp(bin_times, a_times, a_values), "-o", color=dark_a, label="$A'$")
pyplot.plot(bin_times, numpy.interp(bin_times, b_times, b_values), "-o", color=dark_b, label="$B'$")
pyplot.plot(bin_times, numpy.interp(bin_times, c_times, c_values), "-o", color=dark_c, label="$C'$")
pyplot.plot(bin_times, numpy.interp(bin_times, d_times, d_values), "-o", color=dark_d, label="$D'$")
pyplot.legend(loc="best")
pyplot.gcf().set_size_inches(10, 8)
```

A naive piecewise aggregate approximation (PAA) determines the bins that each sample falls into, sums the sample values within each bin, and divides by the number of samples in each bin to get an average value for the bin.  This acts as a low-pass filter that greatly reduces aliasing problems for $D'$ and produces results similar to PLA for $C'$, but note the problems in $B'$ where a bin doesn't contain any samples, causing a divide-by-zero error, a NaN value in the results, and a break in the plot.  Like the first PLA implementation, this algorithm produces NaN values outside the timeseries domain:

```python
def naive_paa(times, values, bin_edges):
    bin_indices = numpy.digitize(times, bin_edges) - 1
    bin_counts = numpy.bincount(bin_indices, minlength=len(bin_edges)-1)
    bin_sums = numpy.bincount(bin_indices, values, minlength=len(bin_edges)-1)
    bin_values = bin_sums / bin_counts
    return bin_values

pyplot.vlines(bin_edges, -1, 1, color=light_bin)
pyplot.plot(a_times, a_values, "-o", color=light_a, markeredgecolor=light_a)
pyplot.plot(b_times, b_values, "-o", color=light_b, markeredgecolor=light_b)
pyplot.plot(c_times, c_values, "-o", color=light_c, markeredgecolor=light_c)
pyplot.plot(d_times, d_values, "-o", color=light_d, markeredgecolor=light_d)

pyplot.plot(bin_times, naive_paa(a_times, a_values, bin_edges), "-o", color=dark_a, label="$A'$")
pyplot.plot(bin_times, naive_paa(b_times, b_values, bin_edges), "-o", color=dark_b, label="$B'$")
pyplot.plot(bin_times, naive_paa(c_times, c_values, bin_edges), "-o", color=dark_c, label="$C'$")
pyplot.plot(bin_times, naive_paa(d_times, d_values, bin_edges), "-o", color=dark_d, label="$D'$")

pyplot.legend(loc="best")
pyplot.gcf().set_size_inches(10, 8)
```
