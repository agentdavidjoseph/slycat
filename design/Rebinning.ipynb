```python
%matplotlib inline
import numpy
import scipy.interpolate
import scipy.signal
import matplotlib
import matplotlib.pyplot as pyplot

matplotlib.rc("lines", markersize=5.0)
lightblue = (0.7, 0.7, 1.0)
lightgreen = (0.7, 1.0, 0.7)
lightred = (1.0, 0.7, 0.7)
lightgray = (0.8, 0.8, 0.8)
```

Create timeseries _a_, _b_, and _c_ which have different start and end times and nonuniform sampling:

```python
a_times = numpy.linspace(0, numpy.pi * 2, 100) ** 3
a_values = numpy.sin(a_times * 0.05)

b_times = numpy.linspace(0, numpy.pi * 1.2, 37) ** 4 + 20
b_values = numpy.sin(b_times * 0.017)

c_times = numpy.linspace(-1, 1, 23)
c_values = scipy.signal.gausspulse(c_times)
c_times = c_times * 100.0 + 121.5
```

Define our resampling bins, using the extents of timeseries _a_ because it has the earliest start time and latest end time:

```python
bin_times, bin_width = numpy.linspace(a_times.min(), a_times.max(), 20, retstep=True)
bin_edges = numpy.append(bin_times - (bin_width / 2), [bin_times[-1] + (bin_width / 2)])

pyplot.vlines(bin_edges, -1, 1, color=lightgray)
pyplot.plot(a_times, a_values, "-o", color="blue", label="a")
pyplot.plot(b_times, b_values, "-o", color="green", label="b")
pyplot.plot(c_times, c_values, "-o", color="red", label="c")
pyplot.legend()
pyplot.gcf().set_size_inches(10, 8)
```

Perform naive resampling on timeseries _a_ by determining the bin indices that each sample falls into, summing the sample values within each bin, and dividing by the number of samples in each bin to get an average:

```python
bin_indices = numpy.digitize(a_times, bin_edges) - 1
bin_counts = numpy.bincount(bin_indices)
bin_sums = numpy.bincount(bin_indices, a_values)
bin_values = bin_sums / bin_counts

pyplot.vlines(bin_edges, -1, 1, color=lightgray)
pyplot.plot(a_times, a_values, "-o", color=lightblue, markeredgecolor=lightblue)
pyplot.plot(bin_times, bin_values, "-o", color="blue")
pyplot.gcf().set_size_inches(10, 8)
```

Perform naive resampling on timeseries _b_ using the same method.  Note that there are problems when a bin doesn't contain any samples - the bin count is zero, causing a divide-by-zero warning and producing a NAN in the results:

```python
bin_indices = numpy.digitize(b_times, bin_edges) - 1
bin_counts = numpy.bincount(bin_indices)
bin_sums = numpy.bincount(bin_indices, b_values)
bin_values = bin_sums / bin_counts

pyplot.vlines(bin_edges, -1, 1, color=lightgray)
pyplot.plot(b_times, b_values, "-o", color=lightgreen, markeredgecolor=lightgreen)
pyplot.plot(bin_times[:len(bin_values)], bin_values, "-o", color="green")
pyplot.gcf().set_size_inches(10, 8)
```

Radial basis function interpolation produces better approximations, including interpolating properly for bins that don't contain samples.  It has extrapolated values outside the domains of timeseries _b_ and _c_.  Also, note the use of a linear basis function for _c_ ... without it, the default multiquadric basis function produces serious ringing.

```python
a = scipy.interpolate.Rbf(a_times, a_values)
b = scipy.interpolate.Rbf(b_times, b_values)
c = scipy.interpolate.Rbf(c_times, c_values, function="linear")

pyplot.vlines(bin_edges, -1, 1, color=lightgray)
pyplot.plot(a_times, a_values, "-o", color=lightblue, markeredgecolor=lightblue)
pyplot.plot(b_times, b_values, "-o", color=lightgreen, markeredgecolor=lightgreen)
pyplot.plot(c_times, c_values, "-o", color=lightred, markeredgecolor=lightred)
pyplot.gcf().set_size_inches(10, 8)

pyplot.plot(bin_times, a(bin_times), "-o", color="blue")
pyplot.plot(bin_times, b(bin_times), "-o", color="green")
pyplot.plot(bin_times, c(bin_times), "-o", color="red")
```

Grid interpolation also yields better results, and returns NaN for times outside the original timeseries domain:

```python
pyplot.vlines(bin_edges, -1, 1, color=lightgray)
pyplot.plot(a_times, a_values, "-o", color=lightblue, markeredgecolor=lightblue)
pyplot.plot(b_times, b_values, "-o", color=lightgreen, markeredgecolor=lightgreen)
pyplot.plot(c_times, c_values, "-o", color=lightred, markeredgecolor=lightred)
pyplot.gcf().set_size_inches(10, 8)

pyplot.plot(bin_times[:len(a_values)], scipy.interpolate.griddata(a_times, a_values, bin_times), "-o", color="blue")
pyplot.plot(bin_times[:len(b_values)], scipy.interpolate.griddata(b_times, b_values, bin_times), "-o", color="green")
pyplot.plot(bin_times[:len(c_values)], scipy.interpolate.griddata(c_times, c_values, bin_times), "-o", color="red")
```
