```python
import numpy
x = numpy.linspace(0, 10, 100)
y = x + numpy.random.normal(size=x.shape)
xy = numpy.column_stack((x, y))
```

```python
import scipy.stats
slope, intercept, r, p, error = scipy.stats.linregress(xy)
```

```python
slope, intercept, r, p, error
```

```python
import toyplot
```

```python
canvas, axes, mark = toyplot.scatterplot(x, y)
axes.plot(numpy.sort(x), numpy.sort(x) * slope + intercept)
```
