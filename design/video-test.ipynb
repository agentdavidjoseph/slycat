```python
import numpy
import toyplot.mp4
```

```python
x = numpy.random.normal(size=100)
y = numpy.random.normal(size=100)
```

```python
canvas, axes, mark = toyplot.scatterplot(x, y, size=64, canvas_style={"background-color":"white"})
def callback(state):
    if state.index() == 0:
        for i in range(len(x)):
            state.set_datum_style(mark, series=0, datum=i, style={"opacity":0.1})
    else:
        state.set_datum_style(mark, series=0, datum=state.index()-1, style={"opacity":1})
canvas.animate(len(x) + 1, callback)
```

```python
toyplot.mp4.render(canvas, "video-test.mp4")
```
